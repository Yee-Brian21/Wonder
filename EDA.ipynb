# Finding the longest time between Assigned job and Accepted job

```python
temp = data[data['Action'] == 'Accepted Job'][['Analyst','Request','Event occurred at']]
```

```python
for r in temp.iterrows():
    temp.at[r[0],'Assigned During'] = data[(data['Action']=='Assigned Job')&(data['Analyst']==r[1]['Analyst'])&(data['Request']==r[1]['Request'])]['Event occurred at'].max()
```

```python
temp.rename(columns={'Event occurred at':'Accepted During'},inplace=True)
```

```python
temp['Wait Time'] = temp['Accepted During']-temp['Assigned During']
```

```python
temp.sort_values('Wait Time')
```

**Longest Wait Time Per Analyst**

```python
temp.groupby('Analyst')['Wait Time'].max().sort_values(ascending=False)
```
